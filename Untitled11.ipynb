```python
df_ml = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_ml['Airline'] = le.fit_transform(df_ml['Airline'])
df_ml['AirportFrom'] = le.fit_transform(df_ml['AirportFrom'])
df_ml['AirportTo'] = le.fit_transform(df_ml['AirportTo'])
```

```python
df_ml.head()
```

```python
corr = df_ml.corr()

corr
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title("Correlation Heatmap")

plt.show()
```

```python
X = df_ml.drop("Class", axis=1)

y = df_ml["Class"]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)

lr.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_lr)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_lr)

cm
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Blues')

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.title("Logistic Regression")

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(
    random_state=42
)
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

dt_accuracy = accuracy_score(y_test, y_pred_dt)

print("Decision Tree Accuracy :", dt_accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.metrics import confusion_matrix

cm_dt = confusion_matrix(y_test, y_pred_dt)

cm_dt
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm_dt,
    annot=True,
    fmt='d',
    cmap='Greens'
)

plt.title("Decision Tree")

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, y_pred_rf)

print("Random Forest Accuracy :", rf_accuracy)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)

cm_rf
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm_rf,
    annot=True,
    fmt='d',
    cmap='Oranges'
)

plt.title("Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
result = pd.DataFrame({

    'Model' : [

        'Logistic Regression',
        'Decision Tree',
        'Random Forest'

    ],

    'Accuracy' : [

        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_dt),
        accuracy_score(y_test, y_pred_rf)

    ]

})

result
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Model',
    y='Accuracy',
    data=result
)

plt.title("Model Accuracy Comparison")

plt.show()
```

```python
importance = pd.DataFrame({

    'Feature' : X.columns,

    'Importance' : rf.feature_importances_

})

importance = importance.sort_values(
    by='Importance',
    ascending=False
)

importance
```

```python
plt.figure(figsize=(8,5))

sns.barplot(

    x='Importance',
    y='Feature',
    data=importance

)

plt.title("Feature Importance")

plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

y_prob = rf.predict_proba(X_test)[:,1]

fpr, tpr, threshold = roc_curve(y_test, y_prob)

roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,6))

plt.plot(fpr, tpr, label="AUC = %0.2f" % roc_auc)

plt.plot([0,1],[0,1],'r--')

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()
```

```python
sample = X_test.iloc[[0]]

prediction = rf.predict(sample)

print("Actual Class :", y_test.iloc[0])

print("Predicted Class :", prediction[0])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {

    "n_estimators":[100,200],

    "max_depth":[10,20,None]

}
```

```python
grid = GridSearchCV(

    RandomForestClassifier(random_state=42),

    param_grid,

    cv=3,

    scoring="accuracy",

    n_jobs=-1

)

grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
best_rf = grid.best_estimator_

best_pred = best_rf.predict(X_test)

print("Optimized Accuracy :")

print(accuracy_score(y_test,best_pred))
```

```python
cm_best = confusion_matrix(y_test,best_pred)

plt.figure(figsize=(6,5))

sns.heatmap(

    cm_best,

    annot=True,

    fmt="d",

    cmap="Greens"

)

plt.title("Optimized Random Forest")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```
